```python
results = []
```

```python
for idx, bus in flix.iterrows():

    route = bus["Route_Number"]
    bus_type = bus["Bus_Type"]
    dep_time = bus["dep_minutes"]
    duration = bus["Journey_Duration_(Min)"]
    date = bus["Departure_Date"]
    price = bus["Weighted_Average_Price"]

    similar = others[
        (others["Route_Number"] == route) &
        (others["Bus_Type"] == bus_type) &
        (others["Departure_Date"] == date) &
        (abs(others["dep_minutes"] - dep_time) <= 60) &
        (abs(others["Journey_Duration_(Min)"] - duration) <= 30)
    ]

    if len(similar) == 0:
        continue

    market_price = similar["Weighted_Average_Price"].mean()
    price_diff = price - market_price

    if price > market_price * 1.2:
        flag = "Overpriced"
    elif price < market_price * 0.8:
        flag = "Underpriced"
    else:
        flag = "OK"

    results.append({
        "Route": route,
        "Date": date,
        "Flixbus_Price": price,
        "Market_Price": market_price,
        "Price_Difference": price_diff,
        "Flag": flag,
        "Competitors_Compared": len(similar)
    })
```

```python
groups = df.groupby(["Route_Number","Bus_Type","Departure_Date"])
```

```python
results = []

for (route, bus_type, date), group in groups:

    flix_group = group[group["Operator"] == "Flixbus"]
    others_group = group[group["Operator"] != "Flixbus"]

    if len(flix_group) == 0 or len(others_group) == 0:
        continue

    for _, bus in flix_group.iterrows():

        dep_time = bus["dep_minutes"]
        duration = bus["Journey_Duration_(Min)"]
        price = bus["Weighted_Average_Price"]

        similar = others_group[
            (abs(others_group["dep_minutes"] - dep_time) <= 60) &
            (abs(others_group["Journey_Duration_(Min)"] - duration) <= 30)
        ]

        if len(similar) == 0:
            continue

        market_price = similar["Weighted_Average_Price"].mean()
        diff = price - market_price

        if price > market_price * 1.2:
            flag = "Overpriced"
        elif price < market_price * 0.8:
            flag = "Underpriced"
        else:
            flag = "OK"

        results.append({
            "Route": route,
            "Date": date,
            "Flixbus_Price": price,
            "Market_Price": market_price,
            "Price_Difference": diff,
            "Flag": flag,
            "Competitors_Compared": len(similar)
        })
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df["Price_Diff_%"] = (
    results_df["Price_Difference"] / results_df["Market_Price"]
) * 100
```

```python
results_df = results_df.sort_values(by="Price_Diff_%", ascending=False)
```
